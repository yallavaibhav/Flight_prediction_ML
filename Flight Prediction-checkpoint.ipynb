```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Converting the Date of Journey into three different parts 
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head(2)
```

```python
df.info()
```

```python
#Converting the object type to integer format

df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head(2)
```

### Arrival time dividing it

```python
#Arrival time 
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_Time'].head()
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minutes'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_minutes'] = df['Arrival_minutes'].astype(int)
```

```python
df.info()
```

```python
#As the Arrival_hour, and Arrival_minutes are made from Arrival_time. So, removing the Arrival_time
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
#Unwanted Route feature
df.drop('Route',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
# Dep_Time dividing the data into parts
df['Dep_hour'] = df['Dep_Time'].str.split(":").str[0]
df['Dep_minutes'] = df['Dep_Time'].str.split(":").str[1]
```

```python
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_minutes'] = df['Dep_minutes'].astype(int)
```
